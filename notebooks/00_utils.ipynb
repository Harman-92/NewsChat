```python
import pygsheets
import pandas as pd
```

```python
SHEET_URL = "https://docs.google.com/spreadsheets/d/1q8wre1CA1a__5O7J3WEb-fC9hcFn2HIiTsrKnHD7o3A"
```

```python
def sheets_to_df(sheet_name, sheet_url):
    """
    Get a Google sheet as a pandas dataframe
    """
    gc = pygsheets.authorize(service_account_file='../google_key.json')
    sh = gc.open_by_url(sheet_url)
    wks = sh.worksheet_by_title(sheet_name)
    return wks.get_as_df()
```

```python
def df_to_sheets(df: pd.DataFrame, sheet_name: str, sheet_url: str):
    """
    Write a pandas dataframe to a Google sheet
    """
    gc = pygsheets.authorize(service_account_file='../google_key.json')
    sh = gc.open_by_url(sheet_url)
    try:
        wks = sh.worksheet_by_title(sheet_name)
    except Exception:
        return f"Sheet: '{sheet_name}' not found"
    wks.set_dataframe(df, (1, 1))
    return f"Data uploaded successfully to sheet: '{sheet_name}'"
```
